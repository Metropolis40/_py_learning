```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
f(10,2)
```

```python
x = np.linspace(0,5,50)
y = np.linspace(0,4,30)

X, Y = np.meshgrid(x,  y)

Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, colors='black');
plt.contour(X, Y, Z, 20, cmap='RdGy');
```
