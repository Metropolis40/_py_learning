### Template 1: draw lines

```python
#Three lines to make our compiler able to draw:
import matplotlib as mpl 
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn')
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])
ypoints2 = np.array([4, 12])

plt.figure( figsize= (12,6))


# here we plot two graphs, we add them as 'layers' to the plate.
plt.plot(xpoints, ypoints, 'o', linestyle = ':', linewidth = '2.5', color = 'k', ms= 10, mec = 'k', mfc = '#fff200' )
plt.plot(xpoints, ypoints2, 'o', linestyle = ':', lw = '2.5', color = 'b', ms= 10, mec = 'k', mfc = 'red' )


font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", 	fontdict = font1, loc = 'left')

plt.xlabel("Average Pulse", 	fontdict = font2)
plt.ylabel("Calorie Burnage", 	fontdict = font2)




plt.grid(axis = 'x')
plt.grid(axis = 'y')
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

#plt.show()

plt.savefig('f0.pdf', dpi=1000)
```

### Template 4: bar charts

```python
#Three lines to make our compiler able to draw:
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize= (12,6))
 


plt.subplot(2,2,1)

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x, y, color = "red", width = 0.5)
#plt.show() 
#plt.show() shows the display but will also clear everything - then we cannot output anything afterwards.
#thus we need to comment out this line so that we can output the graph to an external file.


plt.subplot(2,2,2)

x = np.array(["A1", "v23v", "C", "D"])
y = np.array([23, 81, 21, 3])
plt.barh(x, y, color = "k", height = 0.8)
#plt.show()


plt.subplot(2,2,3)

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x, y, color = "red", width = 0.5)
#plt.show()


plt.subplot(2,2,4)

x = np.array(["A1", "v23v", "C", "D"])
y = np.array([23, 81, 21, 3])
plt.barh(x, y, color = "k", height = 0.8)
#plt.show()

 

plt.savefig('foo.pdf', dpi=1000)
```
